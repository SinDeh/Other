```python
import random
```

Welcome to the Mastermind Game. At first you should enter number's digit, then program generate number randomly with digit you enter. \
for example: number of digit you enter is equal to 4 then program generate number 1234. \
Every time you must guess the 4 digit number. if each number in 4 digit equal to program number, program show that or those number \
and show wrong guessed number with x. for example you guess 3214, program show you x2x4



About winner between program and user :
- Player 1 plays first by setting a multi-digit number.
- Player 2 now tries his first attempt at guessing the number. If Player 2 succeeds in his first attempt 
<br />(despite odds which are highly unlikely) he wins the game and is crowned Mastermind!
<br />If not, then Player 1 hints by revealing which digits or numbers Player 2 got correct.
- The game continues till Player 2 eventually is able to guess the number entirely.
- Now, Player 2 gets to set the number and Player 1 plays the part of guessing the number.
- If Player 1 is able to guess the number within a lesser number of tries than Player 2 took,
<br />then Player 1 wins the game and is crowned Mastermind.
- If not, then Player 2 wins the game.
- The real game, however, has proved aesthetics since the numbers are represented by color-coded buttons.

### User turn to guess program's number

```python
# Get digit of numbers from user for program generate randomly
user_digit_number = int(input(f"How many digit do you want that number have (1 to 10)? "))

# Create list of x for wrong guessed number
star_list = ['x' for i in range(user_digit_number)]

# Range of numbers that program choose randomly from these to generate final number (from 1 to 9)
range_digit = [i for i in range(1, 10)]

# Generate final number with defined count and numbers range 
sys_number = random.sample(range_digit, k= user_digit_number)

# Convert final number from str to list with separating them
sys_number = [str(i) for i in sys_number] # '6912'

# Create count assignment for counting time of user guess
user_count = 0

# Loop work until all is true
while True:

        # At first when program begin. Ask from user to enter his/her first guess
        if user_count == 0:
                user_guess = input(f"Please Guess the {user_digit_number} digit number (by enter word exit you can leave the game): ")
        # It shows to user until last guess for enter guesses
        else:
                user_guess = input(f"Enter your next choice of numbers (by enter word exit you can leave the game):")

        # Whenever user want to leave the game. HE/She just need to enter exit
        if (user_guess.isnumeric() == False) & (user_guess == 'exit'):
                print(f"You decided leave the mastermind game.\nThe true number was:\n\t{' '.join(sys_number)}\nGoodbye until next time.")
                break
        
        # Define some error for guess number input. Such as: All be number, length of number
        if (len(user_guess) != user_digit_number) & (user_guess != 'exit'):
                assert len(user_guess) != user_digit_number
                print(f"Length of number must be {user_digit_number}. Try again.\n")

        elif (user_guess.isnumeric() == False) & (user_guess != 'exit'):
                assert user_guess.isnumeric() == False
                print(f"All letters must be number. Try again.\n")

        # If user enter number as well then program starting certain function on entry number
        else:
                # Count assignment plus 1
                user_count += 1
                
                # Program convert number that user enter, from str to list after separating them
                user_guess = [i for i in ' '.join(user_guess) if i.isnumeric()]
                
                # Find intersection number between number that user guessed and number that program generate it
                same_number = [i for i in user_guess if i in sys_number]

                # Show tries count and guessed number to user 
                print(f"{user_count}th guess ({' '.join(user_guess)}) result is:")
                
                # Loop that processing number element by one by for find same number with same index of program number 
                for i in same_number:
                        for j in range(user_digit_number):
                                if ''.join(user_guess).find(i, j) == ''.join(sys_number).find(i, j) and (''.join(user_guess).find(i, j) >= 0):
                                        star_list[sys_number.index(i)] = i
                                        
                # Show user result of matching guessed number with program number
                print(f"\t{' '.join(star_list)}")
                
        # If user can guess true number in his/her first guess. Program show below sentence
        if (''.join(sys_number) == ''.join(user_guess)) & (user_count == 1):
                print(f"\nAmazing!!!. You rocked it!!!!\nYou've become a Mastermind.\nYou did it in {user_count}th guess.")
                break
        
        # If user guess true number after first guess. Program show below sentence
        elif (''.join(sys_number) == ''.join(user_guess)) & (user_count > 1):
                print(f"\nCongratulation.\nYou've become a Mastermind.\nYou did it in {user_count}th guess.")
                break
        
        # Below sentences show when user guess none or some of the number
        if (star_list.count('x') != user_digit_number) & (''.join(sys_number) != ''.join(user_guess)):
                print(f"True guessed number must be in your next guess.\n")

        elif (star_list.count('x') == user_digit_number) & (''.join(sys_number) != ''.join(user_guess)):
                print(f"None of the numbers in your input match. Try again.\n")
```

### Program turn to guess user's number

```python
# Create list of x for wrong guessed number
star_list = ['x' for i in range(user_digit_number)]

# Create count assignment for counting times of program guess
sys_count = 0

# Loop work until all is true
while True:
    
    # Get number whit certain digit that must be guessed by program, from user
    user_number = input(f"Please enter your {user_digit_number} digit number that program start guessing it: ")

    # Define some error for guess number input. Such as: all be number, length of number
    if (len(user_number) != user_digit_number):
            assert len(user_number) != user_digit_number
            print(f"Length of number must be {user_digit_number}. Try again.\n")

    elif (user_number.isnumeric() == False):
            assert user_number.isnumeric() == False
            print(f"All letters must be number. Try again.\n")
    # If user enter number without problem. Program convert number from str to list after separate them
    else:

        user_number = [i for i in ' '.join(user_number) if i.isnumeric()]
        
        break

# Loop work until all is true
while True:

        # Create set for add index of star list that equal to x (for specific range that program must be guess)
        x_index = set()

        # Work on finding x in star list and add their index to defined set 
        for j in range(user_digit_number):
            if ''.join(star_list).find('x', j) >= 0:
                x_index.add(''.join(star_list).find('x', j))

        # Program start generate numbers randomly by define range
        sys_guess = random.sample(range_digit, k= len(x_index))

        # Convert guessed number to list 
        sys_guess = [str(i) for i in sys_guess]
        
        # Count assignment plus 1
        sys_count += 1
        
        # Show tries count and guessed number to user
        print(f"{sys_count}th guess ({' '.join(sys_guess)}) result is:")

        # Convert set to list because of set not iterable
        x_index = list(x_index)

        # Loop work on add true guessed number with true index to star list
        for i in range(len(x_index)):
            if user_number[x_index[i]] == sys_guess[i]:
                star_list[x_index[i]] = sys_guess[i]
        
        # Show final star list in every count to user
        print(f"{' '.join(star_list)}\n")

        # If program can guess whole number true. Program start Comparing user and program count number for find winner(Mastermind)
        if (''.join(user_number) == ''.join(star_list)):
                        
            if (sys_count < user_count):
                print(f"\nProgram have become a Mastermind.\nProgram did it in {sys_count}th guess.\nAnd You did it in {user_count}th guess.")
                break
            
            elif (sys_count > user_count):
                print(f"\nYou have remain a Mastermind.\nYou did it in {user_count}th guess.\nAnd Program did it in {sys_count}th guess.")
                break
            
            elif (sys_count == user_count):
                print(f"Both, You and program guessed in same chance ({sys_count}).So we don't have mastermind at this time.")
                break
        
```
